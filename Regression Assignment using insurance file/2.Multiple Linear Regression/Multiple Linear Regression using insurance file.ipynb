```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
```

```python
dependent
```

```python
#In this line training test set and test set split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Model creation using train set
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
weight=regressor.coef_
```

```python
weight
```

```python
bais=regressor.intercept_
```

```python
bais
```

```python
#How to evalucate the Model
y_pred=regressor.predict(x_test)
```

```python
#In this line import the r2-score from sklearn libray
from sklearn.metrics import r2_score
#In this line output test variable and output predicted variable assigned into the r-score variable
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#How to save the Model
import pickle
```

```python
#finalized model save using .sav format and assigned into the filename variable
filename="finalized_model_linear.sav"
```

```python
# In this line all file save into the regressor and open the file using write binary
pickle.dump(regressor,open(filename,'wb'))
```

```python
#How to load the Model
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[15,20,25,1,0]])
```
